```python
import xarray as xa
import numpy as np
from pylibs.plot_utils import set_size
import matplotlib.pyplot as plt
```

```python
l64file = '/glade/work/swei/skylab/jedi-bundle/fv3-jedi-data/testinput_tier_1/inputs/fv3files/akbk64.nc4'
l127file = '/glade/work/swei/skylab/jedi-bundle/fv3-jedi-data/testinput_tier_1/inputs/fv3files/akbk127.nc4'
```

```python
l64ds = xa.open_dataset(l64file)
l127ds = xa.open_dataset(l127file)
```

```python
l64pres = l64ds.ak + l64ds.bk * 100000.
l127pres = l127ds.ak + l127ds.bk * 100000.
x64 = np.zeros_like(l64pres)
x64[:] = 0.1
x127 = np.zeros_like(l127pres)
x127[:] = -0.1
```

```python
l127pres.min(), l64pres.min()
```

```python
fig, ax = plt.subplots()
set_size(2, 8)
ax.set_xlim(-0.2, 0.2)
ax.set_yscale('log')
ax.scatter(x64, l64pres, s=20, marker='o', edgecolors='None')
ax.scatter(x127, l127pres, s=20, marker='*', edgecolors='None')
ax.invert_yaxis()
```
